```python
df4ToList = list(df4['filename'])#convert the filename column to a list
df4ToList_no_dup= list(dict.fromkeys(df4ToList)) #supress the duplicate from the list.
```

```python
import os
import pandas as pd
from csv import reader
import csv
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import matplotlib
import matplotlib.pyplot as plt
```

```python
def drawer_from_list5(dir_path):
    try:
        for item in df4ToList_no_dup:
            fig, ax = plt.subplots()
            ax.imshow(Image.open(os.path.join(dir_path, item))) 
            for idx, row in df4.iterrows():
                if item == row["filename"]: 
                    ax.add_patch(patches.Rectangle((row ["X"],row["Y"]),row["Width"],row["Height"], linewidth=1.1, edgecolor='r', facecolor='none')) 
            plt.show()
    except FileNotFoundError as Error:
             print("Error in the file", item)
    
dir_path = "images940" 
drawer_from_list5(dir_path)
```
